```python
df["Project_Name"]=df["Project_Name"].astype(str).str.strip("Project_").astype(int)
```

```python
df["Configuration"]=df["Configuration"].replace({"4bhk":"4BHK",
                                                 "3Bhk":"3BHK",
                                                 "4Bhk":"4BHK",
                                                 "5bhk+":"5BHK+",
                                                 "5Bhk+":"5BHK+",
                                                 "3bhk":"3BHK"})
```

```python
df["Buyer_Comments"] = df["Buyer_Comments"].fillna("No comment")
```

```python
def parse_ticket_price_cr(v):
    """Parse messy ticket price values and return in Crores (float)."""
    if pd.isna(v):
        return np.nan
    s = str(v).replace(",", "").replace("₹", "").strip().lower()
    if s == "":
        return np.nan

    # Regex to capture number + optional unit
    m = re.match(r'^([0-9]+(?:\.[0-9]+)?)\s*([a-zA-Z]*)$', s)
    if m:
        num = float(m.group(1))
        unit = m.group(2)

        # Explicit units
        if "cr" in unit or "crore" in unit:
            return num
        if "lakh" in unit or "lac" in unit:
            return num / 100.0

        # No unit given → apply your rule
        if "." in m.group(1):   # decimal present → Crores
            return num
        else:                   # integer only → Lakhs
            return num / 100.0

    # Fallback: try extracting numbers anyway
    nums = re.findall(r'[\d\.]+', s)
    if not nums:
        return np.nan
    num_str = nums[0]
    num = float(num_str)
    if "." in num_str:
        return num  # decimal → Crores
    else:
        return num / 100.0  # integer → Lakhs

df["Ticket_Price_Cr"] = df["Ticket_Price_Cr"].apply(parse_ticket_price_cr)
df["Ticket_Price_Cr"] = pd.to_numeric(df["Ticket_Price_Cr"], errors="coerce")
```

```python
df["Unit_Size_Sqft"] = pd.to_numeric(df["Unit_Size_Sqft"], errors="coerce")
df.loc[df["Unit_Size_Sqft"] <= 0, "Unit_Size_Sqft"] = np.nan
```

```python
for col in ["Amenity_Score", "Connectivity_Score", "Locality_Infra_Score"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
a=LabelEncoder()
df.loc[:,"NRI_Buyer"] = a.fit_transform(df["NRI_Buyer"])
```

```python
df["Purchase_Quarter"]=pd.to_datetime(df["Purchase_Quarter"],format="%Y-%m-%d")
df["Quarter_Year"] = df["Purchase_Quarter"].dt.to_period("Q").astype(str)
df["Quarter_Number"] = df["Purchase_Quarter"].dt.quarter.astype(int)
df["Purchase_Year"] = df["Purchase_Quarter"].dt.year.astype(int)
```

```python
df["Amenity_Score"] = df.groupby("Micro_Market")["Amenity_Score"]\
                        .transform(lambda x: x.fillna(x.median()))
df["Amenity_Score"] = df["Amenity_Score"].fillna(df["Amenity_Score"].median())
```

```python
df["Unit_Size_Sqft"] = df.groupby("Configuration")["Unit_Size_Sqft"]\
                         .transform(lambda x: x.fillna(x.median()))
df["Unit_Size_Sqft"] = df["Unit_Size_Sqft"].fillna(df["Unit_Size_Sqft"].median())
```

```python
# Step 1: Calculate Price per Sqft (in INR) where both values exist
df["Price_per_Sqft_INR"] = (df["Ticket_Price_Cr"] * 1e7) / df["Unit_Size_Sqft"]

# Step 2: Compute median Price/Sqft for each Micro_Market
pps_market = df.groupby("Micro_Market")["Price_per_Sqft_INR"].median()

# Step 3: Define imputation function
def impute_ticket_price(row):
    if pd.isna(row["Ticket_Price_Cr"]) and pd.notna(row["Unit_Size_Sqft"]):
        # Look up median price per sqft for this market
        market_pps = pps_market.get(row["Micro_Market"], np.nan)
        if pd.notna(market_pps):
            return (row["Unit_Size_Sqft"] * market_pps) / 1e7  # back to Crores
    return row["Ticket_Price_Cr"]

# Step 4: Apply row-wise imputation
df["Ticket_Price_Cr"] = df.apply(impute_ticket_price, axis=1)

# Step 5: Fill any remaining NaN with global median (as last fallback)
df["Ticket_Price_Cr"] = df["Ticket_Price_Cr"].fillna(df["Ticket_Price_Cr"].median())
```

```python
if "Transaction_Type" in df.columns:
    txn = df["Transaction_Type"].astype(str).str.lower()
    df["Booking_Status"] = np.where(txn == "primary", "Booked",
                                    np.where(txn == "secondary", "Resale", "Unknown"))
    df["Booking_Flag"] = np.where(df["Booking_Status"] == "Booked", 1, 0).astype(int)

# Price per sqft in INR: 1 Crore = 10,000,000 INR
df["Price_per_Sqft_INR"] = (df["Ticket_Price_Cr"] * 10_000_000) / df["Unit_Size_Sqft"]
df["Price_per_Sqft_INR"] = df["Price_per_Sqft_INR"].replace([np.inf, -np.inf], np.nan)
df.loc[df["Price_per_Sqft_INR"] <= 0, "Price_per_Sqft_INR"] = np.nan
```
